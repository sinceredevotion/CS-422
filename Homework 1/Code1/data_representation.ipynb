<h1>1. MNIST DATA</h1>

```python
# Read MNIST data
import pandas as pd
import numpy as np

data = pd.read_csv('MNIST_100.csv')
```

```python
# make two variables - X and y
y = data.iloc[:, 0]
X = data.drop('label', axis=1)
#X[0:1]
#np.array(X[0:1])
```

```python
# display the first image
import matplotlib.pyplot as plt
img = np.array(X[900:901]).reshape(28, 28) / 255
plt.imshow(img)
plt.show()
```

```python
# Visualize data
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
PCAX = pca.transform(X)

plt.scatter(PCAX[0:100, 0], PCAX[0:100, 1]) # Digit 0
plt.scatter(PCAX[100:200, 0], PCAX[100:200, 1]) # Digit 1
plt.show()
PCAX
```

<h1>3. Movie data</h1>

```python
import numpy as np
import cv2

cap = cv2.VideoCapture('gangnam.mp4')

while(cap.isOpened()):
    ret, frame = cap.read()
    if not ret:
        break
    #ray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame', np.tile(frame, [1, 2, 1]))
    #cv2.imshow('frame', gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.imshow(frame)
plt.show()
```
